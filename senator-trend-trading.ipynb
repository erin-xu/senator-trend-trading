```python
# IMPORTS
import math
import numpy as np
import pandas as pd
import datetime as dt
import yfinance as yf
from matplotlib import pyplot as plt
#from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_json('https://senate-stock-watcher-data.s3-us-west-2.amazonaws.com/aggregate/all_transactions.json')
print(df.describe())
```

```python
df['disclosure_date'].value_counts()
```

```python
df = df.drop(columns=['owner', 'asset_description', 'asset_type', 'type', 'comment', 'senator', 'ptr_link'])
```

```python
df.info()
```

```python
df = df[df['ticker'] != "N/A"]
df = df[df['ticker'] != "--"]
df
```

```python
stocks = ['AAPL']

data = yf.download(stocks, start="2015-01-01", end="2020-02-21")

print(data)
```

```python
portfolio = []

for row in range(0, df.shape[0]):
    date = df.iloc[row]["transaction_date"]
    month, day, year = date.split("/")
    begin = dt.datetime(int(year), int(month), int(day))
    stop = begin + dt.timedelta(days=1)
    ticker = df.iloc[row]["ticker"]
    try:
        stockinfo = yf.download(ticker, start= begin, end = stop)
        openpos = []
        openpos.append(begin)
        openpos.append(ticker)
        openpos.append(stockinfo.iloc[0]["Open"])
        portfolio.append(openpos)
    except:
        print("delisted")
    
    row += 1

print(portfolio)
```

```python
portfolio = portfolio
```

```python
pf = pd.DataFrame(portfolio)
pf.columns = ['Open Date', 'Ticker', 'Open Price']
pf
```

```python
pf['Open Date'] = pd.to_datetime(pf['Open Date'])
pf.iloc[0]['Open Date']
```

```python
earned = 0
spent = 0
for trade in range(2000, 2200):
    ticker = pf.iloc[trade]['Ticker']
    begin = pf.iloc[trade]['Open Date'].to_pydatetime() + dt.timedelta(days=365)
    stop = begin + dt.timedelta(days=1)
    try:
        stockinfo = yf.download(ticker, start= begin, end = stop)
        earned += (stockinfo.iloc[0]['Open'] - pf.iloc[trade]['Open }Price'])
        spent += (pf.iloc[trade]['Open Price'])
        print(earned)
    except:
        print("no info")
print(earned)
```
